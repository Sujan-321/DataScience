```python
# What is Linear Regression?
    # Linear Regression is a machine learning algorithm based on supervised learning.
    # It is a statical method that is used for predictive analysis.
    # Linear Regression makes predictions for continuous/real or numeric variables such as cost, age, sales, temperature, product's price etc
    # It has 2 types:
        # a. Linear Regression with Single Variable
        # b. Linear Regression with Multiple Variable
    # formula:
        # Linear Equation:    y = mx+c
            # where
                    # y = dependent
                    # x = independent
                    # m = slope/gradient/coefficient
                    # c = intercept
        # Thus, preminum = m*age + c is the formula that we are going to use
```

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv(r"D:\Data_Science\Jupyter_Notebook\MachineLearning\insurance.csv", delimiter='\t')
```

```python
df
```

```python
sns.lmplot(x='age', y='premium', data=df)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['age']], df['premium'])                    # fit is used for training the model, and remember argument/parameter must be in '2D'
```

```python
# now the model is ready for prediction 
```

```python
reg.predict([[21]])   # now, let's check what is the 'premium' should be paid by person whose 'age' is '21'
```

```python
# let's check for age '50'
reg.predict([[50]])
```

```python
# let's go through formula   (y = mx + c)
reg.coef_      # it return slope(m)
```

```python
reg.intercept_    # it return the intercept value(c)
```
