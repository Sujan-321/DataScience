```python
# Degree of Polynomial
# It has 3 types
    # a. 0 Degree Polynomial (y = mx^0 + c = constant)
    # b. 1 Degree Polynomial (y = mx^1 + c = mx+c)
    # c. 2 Degree Polynomial (y = ax^2 + bx + c)

# Generalized Polynomial Equation
    # y = a0 + a1 * X + a2 * X^2 + a3 * X^3 + .... + an * x^n
```

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
df = pd.read_csv(r'Position_Salaries.csv')
```

```python
x = df.iloc[:, 1:2].values  # this is independent variable
```

```python
y = df.iloc[:, 2].values
y
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
# let's go through polynomial regression
# Step 1:
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
```

```python
poly
```

```python
# Step 2:
x_poly = poly.fit_transform(x)
```

```python
# Step 3:
reg2 = linear_model.LinearRegression()
```

```python
# Step 4:
reg2.fit(x_poly, y)
```

```python
# now doing prediction
reg2.predict(poly.fit_transform([[6.5]]))
```
