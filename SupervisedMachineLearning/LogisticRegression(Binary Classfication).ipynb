# What is Logistics Regression?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset 'TermInsuranceDataset'
df_org = pd.read_csv(r'TermInsuranceDataset.csv')
```

```python
df_org
```

# basically, what we gonna do is change 'Bought Insurance' column must be converted into binary form (either '0' or '1')

```python
df = df_org.copy()
```

```python
df['bought_insurance'].replace({'no':'0', 'yes':'1'}, inplace=True)
```

```python
df
```

```python
# now visualize the data using 'scatter plot'
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
plt.scatter(x='age', y='bought_insurance', data=df)
```

```python
# now splitting the dataset(df) in training and testing dataset
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df['bought_insurance'], test_size=0.2)
```

```python
len(x_train)
```

```python
x_train
```

```python
x_test
```

```python
# implement the logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
x_test
```

```python
lr.predict(x_test)  
```

```python
# let's check the '19' year teenagers are interested to buy the 'insurance' or not
lr.predict([[19]])  # if the output is '0'- notinterested, '1' - interested
```
