```python
# creating a function which convert marks into grade points
def grade_points(sub_marks):
    if sub_marks >= 90:
        return 4.0
    elif sub_marks >= 80:
        return 3.6
    elif sub_marks >= 70:
        return 3.2
    elif sub_marks >= 60:
        return 2.8
    elif sub_marks >= 50:
        return 2.4
    elif sub_marks >= 40:
        return 2.0
    elif sub_marks >= 30:
        return 1.8
    elif sub_marks >= 20:
        return 1.6
    elif sub_marks >= 0:
        return 0.8
    else:
        return 0.0
```

```python
# creating a function which converts grade_points into grade
def grade(gp):  # gp is a grade_point
    if gp <= 0.8:
        return "E"
    elif gp <= 1.6:
        return "D"
    elif gp <= 1.8:
        return "D+"
    elif gp <= 2.0:
        return "C"
    elif gp <= 2.4:
        return "C+"
    elif gp <= 2.8:
        return "B"
    elif gp <= 3.2:
        return "B+"
    elif gp <= 3.6:
        return "A"
    elif gp <= 4.0:        
        return "A+"
    else:
        return "F"
```

```python
# apply the above functions from 4th columns, just calculate the gpa_points
for column in df.columns[3:]:
        df[f'{column}_Grade_Point'] = df[column].apply(grade_points)
```

```python
# loading the dataset 
df_gpa = pd.read_csv(r'student_grade_points.csv')
```

```python
df_gpa.head()
```

```python
# Convert GPA columns to numeric, errors='coerce' will convert non-numeric values to NaN
for col in df_gpa.columns[3:]:
    df_gpa[col] = pd.to_numeric(df_gpa[col], errors='coerce')
```

```python
# function that calculate the total gpa
def calculate_total_gpa(row):
    # Select GPA columns, assuming they are from the 4th column onward
    gpa_points = row[3:]  # Skip the first three columns: Name, Gender, DOB
    total_gpa = gpa_points.mean()
    return total_gpa
```

```python
# calculating the total gpa in new column
# Apply the function to calculate GPA for each student
df_gpa['GPA'] = df_gpa.apply(calculate_total_gpa, axis=1)
# df_gpa['GPA'] = df_gpa.apply(lambda row: grade(row['GPA']), axis=1)
```

```python
df_gpa.head()
```

```python
# creating a loop which convert the gpa points into grade
for column in df_gpa.columns[3:-1]:
    df_gpa[f'{column}_gpa'] = df_gpa[column].apply(grade)
```
