```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
sb.get_dataset_names()   # these are the dataset available in the seaborn
```

```python
# lets pick the 'tips' dataset
tipData = sb.load_dataset("tips")
```

```python
tipData.head()
```

```python
sb.barplot(x="day", y="total_bill", hue=tipData.sex, data=tipData) # if we remove the 'data' parameter then we can also write like this 'x=tipData.day',
                                                                        # 'y=tipData.total_bill'
```

```python
orderDay=['Sun', 'Thur', 'Fri', 'Sat']  # if we want to disply particular day data then 
plt.figure(figsize=(16,9))
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay)
plt.show()
```

```python
#if we observe the above graph first show in 'blue' color which represent the 'Male' and 'orange' for 'Female' 
# if we change this order but 'color of male(blue)' assign to the 'female' and vice-versa
hue_order = ['Female', 'Male']
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, hue_order = hue_order)
```

```python
 #in the above chart in the 'y-axis' there is '25' which is the mean of out data, but 
maximum = np.max(tipData['total_bill'])
```

```python
maximum
```

```python
# here maximum data is present is : 50.81, so in this case we use 'estimator' parameter
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, estimator=np.max)
```

```python
# if we customize the size of the line present in the bar by using 'ci(confidence interval)'
```

```python
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, ci=12)
```

```python
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, estimator=np.max, n_boot=2)
```

```python
# if we want to disply the bar in 'horizontal' format then, shift the 'x' and 'y' position
sb.barplot(y='day', x='total_bill', hue='sex', data=tipData, order = orderDay, hue_order = hue_order)
```

```python
# if we use 'orient' parameter to display the chart in horizontal then
sb.barplot(x='total_bill', y='day',hue='sex', data=tipData, order = orderDay, hue_order = hue_order, orient='h')
```

```python
# if we want to change the color then
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, palette=["green", "yellow"])
```

```python
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, saturation=0.3)
```

```python
# if we want to change the color of the line present in the bar use 'errcolor' parameter
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, errcolor='0.5', errwidth=12)
```

```python
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, errcolor='0.5', errwidth=1, capsize=1)
```

```python
# in the chart above data is visualize in two bar in each part then use 'dodge' parameter is use, 'True' is used byDefault
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, order = orderDay, dodge=False)
```

```python
# using 'kwargs' here
kwargs = {"alpha":0.9, "linestyle":":", "linewidth":5, "edgecolor":"k"}
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, **kwargs)
```

```python
# without using **kwargs
```

```python
# kwargs = {"alpha":0.9, "linestyle":":", "linewidth":5, "edgecolor":"k"}
sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, alpha=0.9, linestyle=":", linewidth=5, edgecolor="k")
```

```python
plt.figure(figsize=(16, 9))
ax = sb.barplot(x='day', y='total_bill', hue='sex', data=tipData, alpha=0.9, linestyle="-.", linewidth=5, edgecolor="k")
ax.set(title="BarPlot in tips dataset", xlabel="Days", ylabel="Total Bills")
plt.show()
```
