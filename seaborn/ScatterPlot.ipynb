```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# scatter plot
    # if the two data are present in 'x-axis' and 'y-axis' respectively then we find out the relation between them, known as scatter plot
```

```python
titanicData = sns.load_dataset("titanic")
```

```python
titanicData.head()
```

```python
titanicData
```

```python
sns.scatterplot(x="age", y="fare", data=titanicData)
plt.show()
```

```python
# it looks like here is some problem to take a relation then let's go broader
sns.scatterplot(x="age", y="fare", data=titanicData, hue="sex")
```

```python
sns.scatterplot(x="age", y="fare", data=titanicData, hue="sex", style="who")
```

```python
# if we want to give the different size to them then use 'size' parameter
sns.scatterplot(x="age", y="fare", data=titanicData, hue="sex", size="who", style="who")
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age", y="fare", data=titanicData, hue="sex", size="who", style="who", sizes=(100, 400))
plt.show()
```

```python
sns.scatterplot(x="who", y="fare", data=titanicData, hue="alive", size="who", style="alive")
```

```python
sns.scatterplot(x="who", y="fare", data=titanicData, hue="alive", size="who", style="alive", sizes=(100,400),
               palette="inferno_r", alpha=.7)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age", y="fare", data=titanicData)
sns.lineplot(x="age", y="fare", data=titanicData)
sns.barplot(x="age", y="fare", data=titanicData)
```
