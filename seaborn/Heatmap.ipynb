```python
#Heatmap
# Intro
    # Heatmap is a graphical representation of 2D (two dimensional) data. Each data value represents in a matrix and it has a special color.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arr2D = np.linspace(1, 5, 12).reshape(4, 3)        # here we generate a '12' data that lies between '1 to 5' and represent in '4x3' matrix form
```

```python
arr2D
```

```python
sns.heatmap(arr2D)
```
