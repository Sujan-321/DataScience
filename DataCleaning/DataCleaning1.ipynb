```python
null_var = df.isnull().sum()/df.shape[0] * 100
```

```python
null_var
```

```python
dropColumns = null_var[null_var > 17].keys()  # here we remove the '17%' or more than '17%' null data columns, keys returns the 'columns name'
dropColumns
```

```python
df.columns
```

```python
df1 = df.copy()
```

```python
df1 = df1.drop(columns=dropColumns)
```

```python
df1.shape
```

```python
sns.heatmap(df1.isnull())
```

```python
# then we remove the N/A values from the rows rather than the columns for getting the clean data
cleanDatad2 = df1.dropna()
cleanDatad2.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(cleanDatad2.isnull())
```

```python
cleanDatad2.isnull().sum().sum()    # last sum() represent the total null value present in the dataset
```

```python
# let's do data distributions
cleanDatad2.select_dtypes(include=['int64', 'float64']).columns   # finding the columns whose data types must be in 'int' or 'float'
```

```python
sns.distplot(cleanDatad2['MSSubClass'])
plt.show()
```

```python
# above chart is after data cleaning, but we want to see the before data cleaning
sns.distplot(df['MSSubClass'])
plt.show()
```

```python
# both charts looks like similar but we find out the difference between them, so let's do overlap
sns.distplot(df['MSSubClass'], color='blue', label='Original Data')
sns.distplot(cleanDatad2['MSSubClass'], color='red', label='Cleaned Data')
plt.legend()
plt.show()
```

```python
# now let's do in all columns rather than the single column:
numerical_val = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']
plt.figure(figsize=(20, 20))  # Adjusted size for better fit
for i, var in enumerate(numerical_val):
    plt.subplot(9, 4, i + 1)  # 9 rows x 4 columns
    sns.histplot(df[var], bins=20, color='blue', label='Original', kde=True)
    sns.histplot(cleanDatad2[var], bins=20, color='red', label='Cleaned', kde=True)
    plt.title(var)
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()
```

```python
# now let's perform the 'object' operations
cleanDatad2.select_dtypes(include=['object']).columns   # finding the columns whose data types must be in 'int' or 'float'
# shapeOfCleanData = cleanDatad2.select_dtypes(include=['object']).columns.shape
# shapeOfCleanData
```

```python
pd.concat([df['MSZoning'].value_counts()/df.shape[0]*100,
           cleanDatad2['MSZoning'].value_counts()/cleanDatad2.shape[0]*100], axis=1,
          keys=['MSZoning_org', 'MSZoning_clean']
          )
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
           cleanDatad2[var].value_counts()/cleanDatad2.shape[0]*100], axis=1,
          keys=[var+'_org', var+'_clean'])
```

```python
cat_var_dist("MSZoning")   # our function is perfectly working
```

```python
# now let's do for all variable
variable_value = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
       'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
       'SaleType', 'SaleCondition']

plt.figure(figsize=(20, 20))  # Adjusted size for better fit
for i, var in enumerate(variable_value):
    plt.subplot(10, 4, i + 1)  # 10 rows x 4 columns to fit 37 plots
    sns.countplot(x=var, data=df, color='blue', label='Original', alpha=0.5)
    sns.countplot(x=var, data=cleanDatad2, color='red', label='Cleaned', alpha=0.5)
    plt.title(var)
    plt.xticks(rotation=90)  # Rotate x labels for better readability
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()
```
