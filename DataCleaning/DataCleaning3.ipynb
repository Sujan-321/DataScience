# Numerical Missing Value Imputation By Class

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasetPath = r"https://drive.google.com/uc?export=download&id=1BiGZSedP4BIIuTbVTBodOhVgFImaz08c"
df = pd.read_csv(datasetPath)
```

```python
df.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# here we convert the number into percentage
missingValues = df.isnull().sum()/df.shape[0] * 100
```

```python
miss_val = missingValues[missingValues > 20].keys()
```

```python
df1 = df.copy()
df1 = df1.drop(columns=miss_val)
```

```python
df1.shape
```

```python
df2 = df1.select_dtypes(include=['int64', 'float64'])  # it store numerical values columns and 'object' columns were assign '0'
df2.shape
```

```python
df2.isnull().sum()
```

```python
num_var_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
df2[num_var_miss][df2[num_var_miss].isnull().any(axis=1)]    # df2[num_var_miss].isnull() : This method checks for missing values (NaN) in the selected columns. It returns a DataFrame of the same shape with boolean values, where True indicates a missing value and False indicates a non-missing value.
```

```python
# let's go with 'LotFrontage', if we want to insert the new data in the 'NaN' section then we must have a knowledge about the domain.
# first go with the orginal data
df.head()


# from the below table there are three columns 'LotFrontage' with related about.
# They are :
    # a. LotArea          b. LotShape                c. LotConfig
# we look at the categorical(object) value rather than the numerical(number) value. According to this, we must go through 'LotConfig'
# Note: we predict the 'lotFrontage' NaN value through 'LotConfig' columns
```

```python
# then check the unique values to define the class
df['LotConfig'].unique()
```

```python
# fetching the rows which has 'inside' value in the 'LotConfig' columns
fetch_inside = df[df.loc[:, 'LotConfig'] == "Inside"]["LotFrontage"]
# fetch_fr2 = df[df.loc[:, 'LotConfig'] == "FR2"]["LotFrontage"]
# fetch_corner = df[df.loc[:, 'LotConfig'] == "Corner"]["LotFrontage"]
# fetch_culdsac = df[df.loc[:, 'LotConfig'] == "CulDSac"]["LotFrontage"]
# fetch_fr3 = df[df.loc[:, 'LotConfig'] == "FR3"]["LotFrontage"]

```

```python
fetch_inside.replace(np.nan,fetch_inside.mean())     # here ':' represent the 'rows' having 'Inside' in 'LotConfig' column
```

```python
# here we fill the 'N/A' data with mean 
df_copy = df.copy()
for var_class in df['LotConfig'].unique():
    df_copy.update(df[df.loc[:, 'LotConfig'] == var_class]['LotFrontage'].replace(np.nan,df[df.loc[:, 'LotConfig'] == var_class]['LotFrontage'].mean()))
```

```python
df_copy.isnull().sum()  # data successfully clean
```

```python
# fill all those fields which has 'NaN' values in the whole dataset then we follow the code below:
# here we fill the 'N/A' data with mean 
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:, cat_var] == var_class][num_var_miss].mean()))
```

```python
df_copy[num_vars_miss].isnull().sum()
```

# Data Distribution

# 1. with Mean

```python
df_copy = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:, cat_var] == var_class][num_var_miss].mean()))
```

```python
plt.figure(figsize=(15, 15))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20, kde_kws={"linewidth":5, 'color':'red'}, label="Orginal")
    sns.distplot(df_copy[var], bins=20, kde_kws={"linewidth":3, 'color':'blue'}, label="Mean")
    plt.legend()
```

# 2. with Median

```python
df_copy_median = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig', 'Exterior2nd', 'KitchenQual']

for cat_var, num_var_miss in zip(cat_vars, num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:, cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:, cat_var] == var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize=(15, 15))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20, kde_kws={"linewidth":5, 'color':'red'}, label="Orginal")
    sns.distplot(df_copy[var], bins=20, kde_kws={"linewidth":3, 'color':'blue'}, label="Mean")
    sns.distplot(df_copy_median[var], bins=20, kde_kws={"linewidth":3, 'color':'green'}, label="Median")

    plt.legend()
```

# To check the outlayer

```python
for i, var in enumerate(num_vars_miss):
    plt.figure(figsize=(20, 20))
    plt.subplot(3,1,1)
    sns.boxplot(x=df[var])  # display in horizontal format, that's why we use 'x=' parameter
    plt.subplot(3,1,2)
    sns.boxplot(x=df_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(x=df_copy_median[var])

plt.tight_layout
```
