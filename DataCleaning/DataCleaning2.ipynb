```python
# here we convert the number into percentage
missingValues = df.isnull().sum()/df.shape[0] * 100
```

```python
missingValues
```

```python
# let's remove more than 20% null data from the data set
miss_val = missingValues[missingValues > 20].keys()
```

```python
miss_val
```

```python
# dropping the missing values
df1 = df.copy()
df1 = df1.drop(columns=miss_val)
```

```python
df1.shape
```

```python
df2 = df1.select_dtypes(include=['int64', 'float64'])  # it store numerical values columns and 'object' columns were assign '0'
df2.head()
```

```python
# let's check the missing values using heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(df2.isnull())
```

```python
df2[df2.isnull().any(axis=1)]
```

```python
df2.isnull().sum()
```

```python
missing_num_var = [var for var in df2.columns if df2[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(20, 20))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df2[var], bins=20, kde_kws={"linewidth":5, 'color':'#DC143C'})
```

```python
df2.shape
```

```python
# if the data is in normal distribution then we use 'Mean, Median and Mode' is use to fill those data.
# Normal Distribution : if the chart is bi-symmetrical(if cut in the middle two parts are equal) then is it known as Normal Distribution. For example:
                        # in the above chart 'LotFrontage' is Normal Distrbution


df3 = df2.fillna(df2.mean())   # here we take mean of every columns/variable and assign that mean in the 'N/A'
df3.isnull().sum().sum()
```

```python
plt.figure(figsize=(20, 20))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df2[var], bins=20, kde_kws={"linewidth":5, 'color':'red'}, label="Orginal")
    sns.distplot(df3[var], bins=20, kde_kws={"linewidth":3, 'color':'blue'}, label="Mean")
    plt.legend()
```

```python
df4 = df2.fillna(df2.median())   # here we take mean of every columns/variable and assign that mean in the 'N/A'
df4.isnull().sum().sum()
```

```python
plt.figure(figsize=(20, 20))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df2[var], bins=20, kde_kws={"linewidth":8, 'color':'red'}, label="Orginal")
    sns.distplot(df3[var], bins=20, kde_kws={"linewidth":5, 'color':'green'}, label="Mean")
    sns.distplot(df4[var], bins=20, kde_kws={"linewidth":3, 'color':'blue'}, label="Median")
    plt.legend()
```

```python
# let's discuss about the outlayer
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(20, 20))
    plt.subplot(3,1,1)
    sns.boxplot(x=df[var])  # display in horizontal format, that's why we use 'x=' parameter
    plt.subplot(3,1,2)
    sns.boxplot(x=df3[var])
    plt.subplot(3,1,3)
    sns.boxplot(x=df4[var])

plt.tight_layout()
plt.show()
```

```python
# now let's do concatinate them for more prediction
df_concat = pd.concat([df2[missing_num_var], df3[missing_num_var], df4[missing_num_var]], axis=1)
```
